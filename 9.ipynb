```python
from lib import read_file

import numpy as np
```

```python
lines = read_file("9_input.txt")
lines = [list(map(int, l.split())) for l in lines]
lines
```

```python
total = 0
for l in lines:
    vals = [np.array(l)]
    while not np.all(vals[-1][0] == vals[-1]):
        vals.append(np.diff(vals[-1]))
    
    total += sum(val[-1] for val in vals)

total
```

```python
total = 0
for l in lines:
    vals = [np.array(l)]
    while not np.all(vals[-1][0] == vals[-1]):
        vals.append(np.diff(vals[-1]))
    
    running_val = 0
    for v in vals[::-1]:
        running_val = v[0] - running_val
        
    total += running_val

total
```
