```python
from lib import read_file

from collections import defaultdict
```

```python
def HASH(str):
    tot = 0
    for char in str:
        tot += ord(char)
        tot *= 17
        tot = tot % 256
    
    return tot
```

```python
HASH("HASH")
```

```python
line = read_file("15_input.txt")[0] 

tot = 0
for instr in line.split(","):
    tot += HASH(instr)

tot
```

```python
line = read_file("15_input.txt")[0] 
# line = "rn=1,cm-,qp=3,cm=2,qp-,pc=4,ot=9,ab=5,pc-,pc=6,ot=7"

boxes = defaultdict(dict)
tot = 0
for instr in line.split(","):
    if (idx := instr.find("-")) != -1:
        key = instr[:idx]
        h = HASH(key)
        if key in boxes[h]:
            boxes[h].pop(key)
    else:
        idx = instr.find("=")
        key = instr[:idx]
        h = HASH(key)
        val = int(instr[idx+1:])
        boxes[h][key] = val

boxes

tot = 0
for box in boxes:
    for i, slot in enumerate(boxes[box]):
        tot += (box + 1) * (i + 1) * boxes[box][slot]
        print(tot)

tot
```
