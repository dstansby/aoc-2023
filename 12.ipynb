```python
from lib import read_file

from functools import cache
```

```python
lines = read_file("12_input.txt")
lines
```

```python
@cache
def n_combos(puzzle: str, sizes: list[int]) -> int:

    size_len = len(sizes)
    if puzzle == "" and size_len:
        return 0
    # If less spaces than sizes
    if len(puzzle.replace(".", "")) < sum(sizes):
        return 0
    if sizes == ():
        return len(puzzle.replace("#", "")) == len(puzzle)
    if size_len == 1 and puzzle.replace("?", "") == "":
        return len(puzzle) - sizes[0] + 1
        
    ret = 0
    
    first_size = sizes[0]
    puz_len = len(puzzle)

    # Get location of first #
    first_hash = puzzle.find("#")
    if first_hash == -1:
        first_hash += puz_len

    # Try first size in all positions
    for i in range(0, puz_len - first_size + 1):
        # Make sure no # before current attempt
        if first_hash < i:
            break

        # If next character is a # this size doens't work in this position
        if i + first_size < puz_len and puzzle[i+first_size] == "#":
            continue

        # Get puzzle that's same length as current size
        sub_puzzle = puzzle[i:i+first_size]
        if len(sub_puzzle.replace(".", "")) == len(sub_puzzle):
            # If no . chars, we can fit this size in this sub-puzzle.
            # Drop this size, and search the rest of the puzzle for the
            # rest of the sizes
            ret += n_combos(puzzle[i+first_size+1:], sizes[1:])

    return ret
```

```python
tot = 0
for l in lines:
    puzzle, sizes = l.split()
    sizes = tuple(map(int, sizes.split(",")))
    # print(puzzle, sizes, n_combos(puzzle, sizes))
    tot += n_combos(puzzle=puzzle, sizes=sizes)

tot
```

```python
%%time

tot = 0
for i, l in enumerate(lines):
    puzzle, sizes = l.split()
    sizes = tuple(map(int, sizes.split(",")))

    print(f"{i+1}/{len(lines)}", puzzle, sizes)

    combos = []

    for n in [5]:
        new_puzzle = "?".join([puzzle] * n)
        new_sizes = sizes * n

        new_combos = n_combos(new_puzzle, new_sizes)
        combos.append(new_combos)

    tot += combos[0]

tot
```
